```python
#6. A list represents the ratio between good reviews and bad reviews for each html.
import csv
import re
from itertools import chain
import collections, functools, operator






li=[]
li1=[]
z=0
C=0
with open('short_movies.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        X=row['html'],row['text']
        match= re.findall('(\d+).*(good|fun|funny|great)',str(X))
        match1= re.findall('(\d+).*(bad|sad|suck|sucks)',str(X))
        li.append(match)
        li1.append(match1)
while([] in li) : 
    li.remove([])
while([] in li1) : 
    li1.remove([])
li=list(chain(*li))
li1=list(chain(*li1))

ldict=[]
li = [item for t in li for item in t]
li1 = [item for t in li1 for item in t]
li3=re.findall('(\d+).*?',str(li))
li4=re.findall('(\d+).*?',str(li1))
counter=collections.Counter(li3)
counter1=collections.Counter(li4)
ldict.append({k: counter1[k] / float(counter[k]) for k in counter if k in counter1})
print(ldict)
```

```python
#3. A list of the most frequent 50 words in the file (Uppercase or Lowercase) with the number of their occurrences.
import csv
import re
from itertools import chain
import collections
li=[]
with open('short_tweets.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        X=row['text']
        match= re.findall('[A-Za-z]+',(X))
        li.append(match)

while([] in li) : 
    li.remove([])
li=list(chain(*li))
counter=collections.Counter(li)
print(counter.most_common(50))
```

```python
#4. Which occurs more Love or Hate?!, Which occurs more Good or Bad?!
import csv
import re
from itertools import chain
import collections
li=[]
with open('short_tweets.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        X=row['text']
        match= re.findall('love|hate|good|bad',(X))
        li.append(match)
while([] in li) : 
    li.remove([])
li=list(chain(*li))

counter=collections.Counter(li)
li1=counter.most_common(4)
li1=list(chain(*li1))
i = iter(li1)
b = dict(zip(i, i))

c=b.get('good')
z=b.get('bad')
f=b.get('love')
s=b.get('hate')
key_list = list(b.keys())
val_list = list(b.values())
if c>z:
   print(key_list[val_list.index(c)])
else:
    print(key_list[val_list.index(z)])
if f>s:
   print(key_list[val_list.index(f)])
else:
    print(key_list[val_list.index(s)])
```
